```python
from tkinter import *
from tkinter import ttk, messagebox
from deep_translator import GoogleTranslator

# Translate function
def translate_text():
    try:
        src_lang = src_lang_combo.get()
        dest_lang = dest_lang_combo.get()
        text = text_entry.get("1.0", END).strip()
        
        if not text:
            messagebox.showwarning("Input Error", "Please enter some text to translate")
            return
        
        translated = GoogleTranslator(source=src_lang, target=dest_lang).translate(text)
        result_box.delete("1.0", END)
        result_box.insert(END, translated)
    except Exception as e:
        messagebox.showerror("Error", f"Translation failed: {e}")

# Tkinter Window
root = Tk()
root.title("Simple Language Translator")
root.geometry("600x400")

# Language options
languages = GoogleTranslator().get_supported_languages(as_dict=True)  # {lang: code}
lang_names = list(languages.keys())

# Input Text
Label(root, text="Enter Text:", font=("Arial", 12)).pack()
text_entry = Text(root, height=5, width=70)
text_entry.pack(pady=5)

# Dropdowns for languages
frame = Frame(root)
frame.pack(pady=5)

Label(frame, text="From:").grid(row=0, column=0, padx=5)
src_lang_combo = ttk.Combobox(frame, values=lang_names, width=20)
src_lang_combo.set("auto")  # auto-detect
src_lang_combo.grid(row=0, column=1, padx=5)

Label(frame, text="To:").grid(row=0, column=2, padx=5)
dest_lang_combo = ttk.Combobox(frame, values=lang_names, width=20)
dest_lang_combo.set("english")  # default target language
dest_lang_combo.grid(row=0, column=3, padx=5)

# Translate Button
Button(root, text="Translate", command=translate_text, bg="green", fg="white").pack(pady=10)

# Output Box
Label(root, text="Translated Text:", font=("Arial", 12)).pack()
result_box = Text(root, height=5, width=70)
result_box.pack(pady=5)

root.mainloop()
```
